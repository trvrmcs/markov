As before, but this time the next word is a function not of just the preceding one but the preceding three.

Shannon gets as far as trigrams using purely manual methods!  

Here we generalize to quad-grams

```python
from typing import Sequence, Iterator
from collections import defaultdict, Counter
import random 
from textwrap import fill 

from itertools import islice

rng = random.Random()
```

```python
words=open("../texts/alice.txt").read().split()
```

```python

def quads(sequence: Sequence[str]) -> Iterator[tuple[str, str, str, str]]:
    a, b, c = None, None, None
    for d in sequence:
        if a is not None:
            yield (a, b, c, d)
        a = b
        b = c
        c = d
```

```python
counters = defaultdict(lambda: Counter())

a,b,c=None,None,None
for d in words + words[:4]:
    if a is not None:
        counters[(a, b, c)][d] += 1

    a, b, c = b, c, d


counters = {**counters}
```

```python
counters[("said","to","the")]
```

```python
def process()->Iterator[str]:
    a,b,c = random.choice(list(counters.keys()))
    while True:
        yield c
        counter = counters[(a,b,c)]
        choices, weights = zip(*counter.items())
        a=b
        b=c
        c=rng.choices(choices,weights)[0]
     
```

```python
p=process()
```

```python
for x in islice(process(), 10):
    print(x) 
```

```python
print(fill(" ".join(list(islice(process() ,100)))))
```
