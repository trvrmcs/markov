# Problem 3: Learning

* We have an HMM and we _don't_ know the transition probabilities or observation likelihoods.
* Given an observation sequence, 'learn' these parameters.


We introduce $\beta$, the probability of seeing the observations from time $t+1$ to the end, given H.M.M. $\lambda$ in state $i$ at time $t$.

Definition:


$$
\beta_t(i) = P(o_{t+1}, o_{t+2},...o_T  | q_t = i, \lambda)
$$

Again, we have to compute it recursively, but in this case $\beta_t$ is dependent on $\beta_{t+1}$.

(This is I guess comparable to the back-prop algorithm in neural networks.)



## Algorithm

#### Initialization
$$
\beta_T(i) = 1, 1 \le i \le N
$$

#### Recursion 

$$
\beta_t(i) = \Sigma_{j=1}^N a_{ij} b_j(o_t+1) \beta_{t+1} (j) 
,
 1 \le i \le N,
 i \le t \lt T


$$

#### Termination
$$

P(O|\lambda) = \sigma_{j=1}^N \pi_j b_j(o_1)\beta_1(j)
$$
